```python
from requests_html import HTMLSession
session = HTMLSession()

```

```python
import re
def extract_clean_newsnames(content_html_a):
  # Regular expression pattern to extract titles
  pattern = r"title='(.*?)'"
  # Extract titles using regular expression and store them in a list
  titles = [re.search(pattern, str(item)).group(1) for item in content_html_a if re.search(pattern, str(item))]
  titles_cl = []
  for ttl in titles:
    ttlis = str(ttl).split(' ')
    if len(ttlis) < 3:
      titles_cl.append(ttl)
  clean_title = []
  for cl in titles_cl:
    if cl not in clean_title:
      if ('United Kingdom' not in cl) & ('Print circulation' not in cl) & ('Wikipedia:Citation' not in cl) & ('Wikipedia' not in cl) & (':' not in cl):
        clean_title.append(cl)

  return clean_title
```

```python
def get_all_content_titles(list_sidecontent):
  # connect to needed webpage
  list_cl_titles = []
  root_url = "https://en.wikipedia.org/wiki/List_of_newspapers_in_the_United_Kingdom_by_circulation"
  for xur in list_sidecontent:
    uxrl = root_url + "#" + xur
    resp2 = session.get(uxrl)
    content_ht = resp2.html.find('a')
    clist = extract_clean_newsnames(content_ht)
    for item in clist:
      list_cl_titles.append(item)
  return list_cl_titles
```

```python
all_titles = get_all_content_titles(list_of_sidebarcontent)
```

```python
searchresultlist = []
count = 0
for q in all_titles:
  query = str(q)
  for j in search(query, tld="com", num=1, stop=1, pause=2):
    print(j)
    searchresultlist.append(j)
  count = count + 1
  print(count)
```

```python
init_clean_list = []
for i in searchresultlist:
  if i not in init_clean_list:
    init_clean_list.append(i)
```

```python
# Remove URLs containing "wikipedia.org" and empty strings
cl_url_init = [url for url in init_clean_list if not re.search(r'wikipedia\.org', url) and url]
```

```python
import re


# Create a dictionary mapping original URL to normalized URL
urlnorm_dict = {}
for url in cl_url_init:
    unilist = []
    dict1 = {"cleaned_url":url}
    match = re.search(r'(https://[^/]+/)', url)
    if match:
        dict2 = {"normalized_url":match.group(1)}
    unilist.append(dict1)
    unilist.append(dict2)
    urlnorm_dict[url] = unilist
```
