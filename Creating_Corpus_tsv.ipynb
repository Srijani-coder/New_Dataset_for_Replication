```python
import pandas as pd

# Step 1: Read the text file and create a dictionary
data_dict = {}

with open("/content/sample_data/url_info_biasfact.txt", "r") as file:
    lines = file.readlines()
    lines
```

```python
print(lines)
```

```python
splitted_text = []
for line in lines:
  list_line = line.split(" - ")
  splitted_text.append(list_line)
```

```python
for split in splitted_text:
  count = len(split)
  if count == 2:
    print(split)
```

```python
splitted_text[0]
```

```python
import re
def get_each_line_dict(split):
  # Define the regular expression pattern
  pattern = r'\d+\.'
  # Find all matches in the string
  matches = re.findall(pattern, str(split[0]))
  # Remove the matches from the string
  output_string = re.sub(pattern, '',str(split[0]))

  url = output_string.replace(" ",'')
  print(url)
  bias = split[1]
  fact = split[2].replace("\n",'')

  normalized_url = url.replace("https://", "").replace("www.", "")
  print(normalized_url)
  ref = "https://mediabiasfactcheck.com/" + normalized_url.replace(".com", "")
  print(ref)
  return {"source_url": url, "Normalized_url": normalized_url, "fact" : fact ,  "bias" : bias, "reference" : ref}
```

```python
get_each_line_dict(splitted_text[0])
```

```python
data_dict  = []
for split in splitted_text:
  line_dict = get_each_line_dict(split)
  data_dict.append(line_dict)
```

```python
import csv

# Specify the output file path
output_file = 'corpus.tsv'

# Extract column names from the first dictionary
column_names = data_dict[0].keys()

# Write the data to a TSV file
with open(output_file, mode='w+', newline='') as tsv_file:
    writer = csv.DictWriter(tsv_file, fieldnames=column_names, delimiter='\t')

    # Write the header
    writer.writeheader()

    # Write the data rows
    for row in data_dict:
        writer.writerow(row)

print(f"Data has been written to {output_file}.")
```
