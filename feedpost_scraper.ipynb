I have collected some urls of feedpost sites which have listed blogs,news, media sites about different subjects. Here, I am storing them in a list.

```python
data_collection_feedpost_urls = ['https://journalists.feedspot.com/sports_news_websites/','https://blog.feedspot.com/world_news_blogs/','https://journalists.feedspot.com/marketing_news_websites/',
                                 'https://journalists.feedspot.com/tech_news_websites/','https://tech.feedspot.com/technology_blogs/',
                                 'https://au.feedspot.com/australian_beauty_blogs/','https://books.feedspot.com/literary_blogs/','https://politics.feedspot.com/political_blogs/'
                                 'https://rss.feedspot.com/world_news_rss_feeds/',]
```

```python
# Libraries installed for scraping
from requests_html import HTMLSession
import re
import pandas as pd
# establish a session
session = HTMLSession()
```

```python
# By observing url patterns, I defined a string matching patterns for targeted news media urls listed in sites
# From them, getting news, blogs original urls extracted.
def scrape_and_load_news_urls(feed_posturl):
  resp = session.get(feed_posturl)
  links = resp.html.find("a")
  pattern = 'https://([^/]+).([^/]+).([^/]+)/?utm_source=feedspot'
  urls = []
  for l in links:
    l = str(l)
    match_l = re.search(pattern,l)
    if match_l:
      res = str(match_l.group(0)) + str(match_l.group(1)) + str(match_l.group(2))
      res_l = res.split('?')
      print(res_l[0])
      urls.append(res_l[0])
  # Writing all extracted urls in text file
  with open('/content/sample_data/cleaned_news_urls.txt','w') as file:
    for ur in urls:
      file.write(ur +'\n')
```

```python
# Now applying the function and scrape target data
for url in data_collection_feedpost_urls:
  scrape_and_load_news_urls(url)
```
